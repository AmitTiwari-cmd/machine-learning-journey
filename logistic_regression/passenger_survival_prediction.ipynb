```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
raw_df = pd.read_csv(r"C:\Users\Amit Tiwari\OneDrive\Desktop\python\kaggle\Titanic-Dataset.csv")
raw_df.head()
```

```python
new_df = raw_df[["Pclass", "Sex", "Age", "Fare", "SibSp", "Parch"]]
new_df.Age = new_df.Age.fillna(new_df.Age.mean()).round()
new_df.head()
sex_dict = { 
    "female" : 0,
    "male" : 1
}
new_df["Sex"] = new_df["Sex"].map(sex_dict)
result_series = raw_df["Survived"]
```

```python
train_data,test_data,train_result,test_result = train_test_split(new_df,result_series,test_size=0.2,random_state=42)
model = LogisticRegression()
model.fit(train_data,train_result)
prediction = model.predict(test_data)
acc_score = accuracy_score(test_result,prediction)
print(f"Accracy Score : {acc_score:.2f}")
```

```python
new_passenger = [[1, 1, 25, 100, 0, 0]]
will_survive = model.predict(new_passenger)
prob = model.predict_proba(new_passenger)
print(f"Probability of Survival : {prob.round(2)}")
print(f"If new passenger Survive? : {will_survive[0]}")
```

```python
coeff = model.coef_[0].tolist()
print(type(test_data.columns.tolist()))
print(test_data.columns.tolist())
dict_coeff = {
    "Features" : test_data.columns,
    "Coefficient" : model.coef_[0].tolist()
}
print(dict_coeff)
df_coeff = pd.DataFrame(dict_coeff)
print(df_coeff)
```
